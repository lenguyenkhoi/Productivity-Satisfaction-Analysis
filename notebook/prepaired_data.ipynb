```python
import pandas as pd
```

```python
data = pd.read_csv("data/raw/hr_dashboard_data.csv")
data.head()
```

```python
data.info()
```

```python
data["Joining Date"]
```

```python
data_date = data.copy()
data_date['Joining Date'] = pd.to_datetime(data_date['Joining Date'], format='%b-%y')

```

```python
data_date.head()
```

```python
data_date["Joining Date"].value_counts().sort_index()
```

```python
data_salary = data_date.copy()
data_salary["Salary"].isnull().sum()
```

```python
data_salary["Salary"] = data_salary["Salary"].fillna(data_salary["Salary"].median())
data_salary.info
```

```python
data_salary.head()
```

```python
data_dup = data_salary.copy()
data_dup = data_dup.drop_duplicates(subset="Name")
data_dup.info()
```

```python
# Experience Years (từ Joining Date đến hiện tại)
data_year = data_salary.copy()

today = pd.Timestamp.today()
today
```

```python
data_year['Experience Years'] = (today - data_year['Joining Date']).dt.days / 365
data_year['Experience Years'] = data_year['Experience Years'].round(0)
```

```python
data_year["Experience Years"].value_counts().sort_index()
```

```python
data_year.head()
```

```python
# Productivity Group (Low: <50%, Medium: 50-80%, High: >80%)
data_product = data_year.copy()

# hàm product 
def productivity_group(product):
    if product < 50:
        return "Low"
    elif 50 <= product <= 80:
        return "Medium"
    else:
        return "High"

data_product["Productivity Group"] = data_product["Productivity (%)"].apply(productivity_group)
 
```

```python
data_product.head(10)
```

```python
data_product.to_csv("data/processed/hr_dashboard_data_clean.csv")
```
